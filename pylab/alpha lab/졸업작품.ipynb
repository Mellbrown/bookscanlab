```python
# 이미지 저장

import cv2
import numpy as np

cap = cv2.VideoCapture('http://192.168.0.43:4747/mjpegfeed')

while True:
    retval, src = cap.read()
    if not retval:
        break
    
    rotated = cv2.rotate(src, cv2.ROTATE_90_CLOCKWISE) # 회전
    gray = cv2.cvtColor(rotated, cv2.COLOR_BGR2GRAY) # 그레이 스케일
    cv2.imshow('gray', gray)
    edge = cv2.Canny(rotated, 200, 250) # 캐니값
    cv2.imshow('edge', edge)
    
    key = cv2.waitKey(25)
    if key == 27:
        break

cv2.imwrite('./gray.bmp', gray)
cv2.imwrite('./edge.bmp', edge)

if cap.isOpened():
    cap.release()

cv2.destroyAllWindows()
```

```python
# 트랙바 안드로이드 캐니 엣지
# 일단 캐니 엣지 30만으로도 충분해보인다

import cv2
import numpy as np

cap = cv2.VideoCapture('http://172.30.1.32:4747/mjpegfeed')
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
              int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print('frame_size=',frame_size)

retval, frame = cap.read()
frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow('img',frame)

threshold, diff = (0,0)

def onChange(pos):
    global threshold, diff
    threshold = cv2.getTrackbarPos('threshold', 'img')
    diff = cv2.getTrackbarPos('diff', 'img')
    
cv2.createTrackbar('threshold', 'img', 0, 255, onChange)
cv2.createTrackbar('diff', 'img', 0, 255, onChange)

while True:
    retval, frame = cap.read()
    if not retval:
        break
    frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
    cv2.imshow('img', cv2.Canny(frame, threshold, threshold, diff))
    
    key = cv2.waitKey(25)
    if key == 27:
        break

if cap.isOpened():
    cap.release()
cv2.destroyAllWindows()
```

```python
# 트랙바 안드로이드 카메라 융합 색깔 분해

import cv2
import numpy as np

cap = cv2.VideoCapture('http://192.168.0.43:4747/mjpegfeed')
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
              int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print('frame_size=',frame_size)

retval, frame = cap.read()
cv2.imshow('img',frame)

img = np.zeros((512, 512, 3), np.uint8)

def onChange(pos):
    global r, g, b
    r = cv2.getTrackbarPos('R', 'img') / 100
    g = cv2.getTrackbarPos('G', 'img') / 100
    b = cv2.getTrackbarPos('B', 'img') / 100
    
cv2.createTrackbar('R', 'img', 0, 100, onChange)
cv2.createTrackbar('G', 'img', 0, 100, onChange)
cv2.createTrackbar('B', 'img', 0, 100, onChange)

font = cv2.FONT_HERSHEY_SIMPLEX
org = (10, 50)

while True:
    retval, frame = cap.read()
    if not retval:
        break
    sb, sg, sr = cv2.split(frame)
    dst = cv2.merge([sb * b, sg * g, sr * r])
    cv2.putText(dst, 'r: %d, g: %d, b:%d' % (dst[50,50,2], dst[50,50,1] ,dst[50,50,0]), org, font,1, (0,255,0), 2)
    cv2.imshow('img', dst)
    
    key = cv2.waitKey(25)
    if key == 27:
        break

if cap.isOpened():
    cap.release()
cv2.destroyAllWindows()
```

```python
# 트랙바 이벤트 처리
import cv2
import numpy as np

def onChange(pos):
    global img
    r = cv2.getTrackbarPos('R', 'img')
    g = cv2.getTrackbarPos('G', 'img')
    b = cv2.getTrackbarPos('B', 'img')
    img[:] = (b, g, r)
    cv2.imshow('img', img)
    
img = np.zeros((512, 512, 3), np.uint8)
cv2.imshow('img', img)
    
cv2.createTrackbar('R', 'img', 0, 255, onChange)
cv2.createTrackbar('G', 'img', 0, 255, onChange)
cv2.createTrackbar('B', 'img', 0, 255, onChange)

cv2.setTrackbarPos('B', 'img', 255)

cv2.waitKey()
cv2.destroyAllWindows()
```
